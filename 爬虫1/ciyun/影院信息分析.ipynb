```python
import numpy as np
import pandas as pd
from wordcloud import WordCloud
import jieba
import matplotlib.pyplot as plt
```

```python
# 读取文件
text = open('12.txt','r',encoding='utf-8').read()
```

```python
# 分割歌词
cut_text = jieba.cut(text)
print(cut_text)
```

```python
# 拼接
result = " ".join(cut_text)
print(result)
```

```python
# 生成词云
wordc = WordCloud(font_path='simhei.ttf',
                 background_color='white',
                 width = 100,
                 height=50,
                 max_font_size=50,
                 min_font_size=10,
                  max_words=1000,
                 mask=plt.imread("0.jpg")
    
                 )
wordc.generate(result)
wordc.to_file('jielun.png')
```

```python
# 显示图片
plt.figure('致橡树')
plt.imshow(wordc)
plt.axis('off')
plt.show()
```
